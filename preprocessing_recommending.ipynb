```python
import pandas as pd
import nltk
```

```python
df = pd.read_pickle("./pickled_file/clean_dataset.pkl")
df.reset_index(inplace=True,drop=True)
df['fullplot'] = df['plot'].values
df.drop('plot',axis=1,inplace=True)
df.head(3)
```

```python
df["director"] = df["director"].apply(lambda x: x.split(','))
df["genre"] = df["genre"].apply(lambda x: x.split(','))
df["cast"] = df["cast"].apply(lambda x: x.split(',')[:3])
df["language"] = df["language"].apply(lambda x: x.split(',')[:3])
df["country"] = df["country"].apply(lambda x: x.split(',')[:3])
df["type"] = df["type"].apply(lambda x: x.split(' '))

```

```python
df.head(2)
```

```python
def change(data):
    l = []
    for i in data:
        i = i.lower()
        i = i.replace(" ","")
        l.append(i)
    return l
```

```python
df["director"] = df["director"].apply(change)
df["cast"] = df["cast"].apply(change)
df["genre"] = df["genre"].apply(change)
df["type"] = df["type"].apply(change)
df["language"] = df["language"].apply(change)
df["country"] = df["country"].apply(change)
```

```python
df.head(2)
```

```python
df["year"]  =df["year"].apply(lambda x: int(x)-int(x)%10)
df["year"] = df["year"].apply(lambda x : str(x).split(','))

df.head()
```

```python
df["movie_tags"]  = df["year"] + df["genre"] + df["director"] + df["cast"] + df["language"] + df["country"] + df["type"]
df["movie_tags"]
```

```python
reviews = list(df["fullplot"].values)
reviews
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import string




def preprocess_sentences(sentences):
    
    sentences = [''.join([char for char in sentence if char not in string.punctuation]) for sentence in sentences]
    
    
    tokenized_sentences = [nltk.word_tokenize(sentence.lower()) for sentence in sentences]
    
    
    english_stopwords = set(stopwords.words('english'))
    
    
    processed_sentences = [[word for word in words if word not in english_stopwords ] for words in tokenized_sentences]
    
    
    stemmer = PorterStemmer()
    processed_sentences = [[stemmer.stem(word) for word in words] for words in processed_sentences]
    
    return processed_sentences

preprocessed =preprocess_sentences(reviews)  
```

```python
df["review_tags"] = preprocessed
```

```python
df["all_tags"] = df["review_tags"] + df["movie_tags"]
df["all_tags"] = df["all_tags"].apply(lambda x: " ".join(x))
```

```python
new_df2 = df[['imdbID','title','movie_tags','review_tags','all_tags']]
```

```python
new_df2["movie_tags"] = new_df2["movie_tags"].apply(lambda x: " ".join(x))
new_df2["review_tags"] = new_df2["review_tags"].apply(lambda x: " ".join(x))
```

```python
new_df2.head(2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer( max_features=20000)
feature_array1 = tfidf_vectorizer.fit_transform(new_df2["movie_tags"]).toarray()
tfidf_vectorizer1 = TfidfVectorizer( max_features=10000,ngram_range=(2,3))
feature_array2 = tfidf_vectorizer1.fit_transform(new_df2["review_tags"]).toarray()
```

```python
import numpy as np
joined_array = np.concatenate((feature_array1,feature_array2), axis=1)
joined_array = joined_array.astype(np.float16)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similartiy = cosine_similarity(joined_array)

similartiy = similartiy.astype(np.float16)
```

```python
new_df2['imdbID'] = new_df2['imdbID'].astype(str).str.zfill(7)
```

```python
def recommend_movie(movie):
    movie_index = new_df2[new_df2['title']==movie].index[0]
    distance = similartiy[movie_index]
    movie_list = sorted(enumerate(distance),key=lambda x: x[1],reverse=True)[1:6]
    for i in movie_list:
        print(new_df2.iloc[i[0],1],new_df2.iloc[i[0],0])
```

```python
recommend_movie("Batman Begins")
```
